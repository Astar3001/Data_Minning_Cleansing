# Menampilkan Data

```python
# Definisikan fungsi format
def format_no_hp(no_hp):
    if pd.isna(no_hp):
        return None
    no_hp = str(no_hp).strip().replace("'", "")  # buang tanda kutip kalau ada
    if no_hp.startswith("62"):
        return "+" + no_hp
    elif no_hp.startswith("0"):
        return "+62" + no_hp[1:]
    else:
        return "+62" + no_hp  # fallback kalau tidak ada awalan

# Terapkan fungsi ke kolom No HP
data["No HP"] = data["No HP"].apply(format_no_hp)

# Lihat hasil
print(data.head())
```

## Memisahkan Nama dan Email

```python
# Function to parse the 'Nama' column and extract name, email, and other info
def parse_nama_column(row):
    items = [item.strip() for item in str(row['Nama']).split(',')]
    name = items[0]
    email = None
    other_info = []

    for item in items[1:]:
        # Simple check for email format (contains '@' and '.')
        if '@' in item and '.' in item:
            email = item
        else:
            other_info.append(item)

    return name, email, ', '.join(other_info)

# Apply the parsing function to each row
data[['Nama', 'Email_extracted', 'Other_info']] = data.apply(parse_nama_column, axis=1, result_type='expand')

# Update the 'Email' column if it's currently NaN and an email was extracted
data['Email'] = data['Email'].fillna(data['Email_extracted'])

# Drop the temporary 'Email_extracted' column
data = data.drop(columns=['Email_extracted'])

# Display the updated DataFrame head and info to see the changes
display(data.head())
display(data.info())
```

# Memisahkan dan menambahkan alasan kompain dan tanggal

```python
import re

def extract_complaint_and_date(row):
    other_info = str(row['Other_info']) if not pd.isna(row['Other_info']) else ""
    complaint = row['Alasan Komplain']
    date = row['Tanggal']

    # Look for complaint reasons and dates in 'Other_info' if 'Alasan Komplain' or 'Tanggal' are missing
    if pd.isna(complaint) or pd.isna(date):
        items = [item.strip() for item in other_info.split(',')]
        for item in items:
            # Simple check for date format (YYYY-MM-DD)
            date_match = re.search(r'\d{4}-\d{2}-\d{2}', item)
            if date_match and pd.isna(date):
                try:
                    date = pd.to_datetime(date_match.group(0))
                except:
                    pass # Ignore invalid date formats
            elif pd.isna(complaint):
                 # Assume any other non-date item is a complaint reason if complaint is missing
                if not date_match and item and not item.startswith(('+62', '0', '(', 'M.', 'S.', 'dr.', 'Ir.', 'H.', 'Hj.')): # Avoid extracting phone numbers or titles
                     complaint = item

    return complaint, date

# Apply the extraction function to update 'Alasan Komplain' and 'Tanggal'
data[['Alasan Komplain', 'Tanggal']] = data.apply(extract_complaint_and_date, axis=1, result_type='expand')

# Display the updated DataFrame head and info to see the changes
display(data.head())
display(data.info())
```

# Memisahkan nomer telpon dari colom tanggal

```python
import re

def clean_phone_number(row):
    phone_number = row['No HP']
    other_info = str(row['Other_info']) if not pd.isna(row['Other_info']) else ""

    # Prioritize existing 'No HP' if it exists and looks like a phone number
    if pd.notna(phone_number) and re.search(r'\d', str(phone_number)):
        # Clean and format the existing phone number
        cleaned_number = re.sub(r'\D', '', str(phone_number)) # Remove non-digits
        if cleaned_number.startswith("62"):
            return "+" + cleaned_number
        elif cleaned_number.startswith("0"):
            return "+62" + cleaned_number[1:]
        elif len(cleaned_number) > 5: # Assume it's a local number or missing prefix if long enough
             return "+62" + cleaned_number # Default to +62
        else:
             return None # Not a valid phone number

    # If 'No HP' is missing or not a valid number, try to extract from 'Other_info'
    if other_info:
        # Look for common phone number patterns in 'Other_info'
        phone_matches = re.findall(r'\+?\d[\d\s\(\)-]{5,}\d', other_info) # More flexible pattern

        for match in phone_matches:
            cleaned_number = re.sub(r'\D', '', match) # Remove non-digits
            if cleaned_number.startswith("62"):
                return "+" + cleaned_number
            elif cleaned_number.startswith("0"):
                return "+62" + cleaned_number[1:]
            elif len(cleaned_number) > 5: # Assume it's a local number or missing prefix if long enough
                 return "+62" + cleaned_number # Default to +62

    return None # Return None if no valid phone number is found

# Apply the cleaning function to each row
data['No HP'] = data.apply(clean_phone_number, axis=1)

# Display the updated DataFrame head and info to see the changes
display(data.head())
display(data.info())
```
